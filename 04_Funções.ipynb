# **Introdução**

Em programação, uma função é uma sequência nomeada de instruções que executa uma operação de computador. Ao definir uma função, especifica-se: i) o nome; e a ii) sequência de instruções. Depois disso, é possível chamá-la pelo nome quantas vezes forem necessárias.

## **Chamada de Função**

Já vimos até aqui diversos exemplos de chamadas de [funções embutidas](https://docs.python.org/pt-br/3/library/functions.html#built-in-functions), como por exemplo a função [*input*](https://docs.python.org/pt-br/3/library/functions.html#input):

```python
nome = input('Digite seu nome: ')
```

Quando fazemos isso, o terminal ou prompt de comando irá:

- Mostrar o texto "Digite seu nome: " na tela; e
- Aguardar até que o usuário digite alguma informação.

A função para imprimir dados em Python é a função [*print*](https://docs.python.org/pt-br/3/library/functions.html#print). Ela é responsável por mostrar valores em seu terminal:



```python
print('Olá', nome)
```

Em qualquer função, a expressão entre parênteses é chamada de argumento. É comum dizer que uma função recebe um argumento e retorna um resultado. O resultado também é chamado de valor de retorno. Para a primeira função, o resultado é uma string com o que o usuário digitou. Embora a segunda função imprima dados na tela, a mesma não retorna um valor, ou melhor, retorna **None**. Observe abaixo:

```python
retorno = print('Olá', nome)
print(retorno)
```

**Atenção:** o objeto Python None denota falta de valor. 

```python
type(retorno)
```

## **Parâmetros de função**

São variáveis declaradas no cabeçalho da função, com uso exclusivo dentro do bloco de instrução da mesma. A definição dos parâmetros que uma função deve receber obriga o envio dos valores a parâmetros sem valores *default*, do contrário, a função não é invocada.

### **O parâmetro sep da função *print***

Por padrão, quando utilizamos virgula para separar os itens, a função *print* utiliza espaços para separar cada saída. Porém, podemos utilizar o parâmetro *sep* para definir um caractere de separação. Entenda no exemplo a seguir:

```python
print('Dia', 'Mês', 'Ano')
```

```python
print('Dia', 'Mês', 'Ano', sep='/')
print('ontem', 'Hoje', 'Amanhã', sep='-')
print("B", "n", "n", ".", sep='a')
```

### **O parâmetro *end* da função *print***

Por padrão, a função *print* utiliza a quebra de linha '\n' como último caracter. O parâmetro end é responsável por alterar esse comportamento, possibilitando ao desenvolvedor trocar qual caracter será adicionado ao final do dado impresso no terminal. Observe:

```python
# Exemplo com fim de linha sem nenhum caracter
print('Vamos estudar ', end='')
print('Python')

# Exemplo com fim de linha igual à ->
print('Quando ensino', end=' -> ')
print('Aprendo')

# Exemplo com fim de linha igual à :
print("Nota", end=': ')
print(10)
```

### **Parâmetro *file* da função *print***

A função *print* também funciona para gravar dados em arquivos. Para isso, utilizamos o parâmetro *file* da função print. Para tanto, precisamos de um arquivo aberto, o que é feito utilizando-se a função open. Veja o exemplo abaixo:

```python
with open('arquivo.txt', 'w') as arquivo:
    print('Primeira linha do arquivo', file=arquivo)
    print(' Segunda linha do arquivo', file=arquivo)
```

- Para abrir um arquivo, o *Python* possui a função *open*. Ela recebe dois parâmetros: o primeiro é o nome do arquivo a ser aberto, e o segundo parâmetro é o modo que queremos trabalhar com esse arquivo - se queremos ler ou escrever. O modo é passado através de uma string: 'w' (abreviação para write) para escrita e 'r' (abreviação para read) para leitura, 'a' (abreviação para appending) para anexar.

- A instrução *with* é utilizada para garantir a finalização de recursos adquiridos. Contudo, também é possível utilizar utilizar as instruções *try* e *finally*, conforme exemplo abaixo, para abrir o arquivo, escrever e encerrá-lo. 



```python
try:
  arquivo = open('arquivo.txt', 'a')
  print("Terceira linha do arquivo", file=arquivo)
  print("Quarta linha do arquivo", file=arquivo)
finally:
  arquivo.close()
```

# **Como Acrescentar Novas Funções?**

Até agora usamos funções que vêm do *Python*, mas também é possível acrescentar novas funções. Uma definição de função especifica o nome de uma nova função e a sequência de instruções que são executadas quando a função é chamada. Abaixo está um exemplo:

```python
def print_refrao():
  print('\nSalve, ó terra dos altos coqueiros')
  print('De belezas soberbo estendal')
  print('Nova Roma de bravos guerreiros')
  print('Pernambuco imortal, imortal!\n')
```

**def** é uma palavra-chave que indica uma definição de função. O nome da função é print_refrao. As regras para os nomes da função são as mesmas que as das variáveis: letras, números e sublinhados são legais, mas o primeiro caractere não pode ser um número.

**Atenção:** não podemos usar palavras-chave como nome de uma função e devemos evitar ter uma variável e uma função com mesmo nome.

Os parênteses vazios depois do nome indicam que esta função não usa argumentos. A primeira linha da definição de função chama-se cabeçalho; o resto é chamado de corpo. O cabeçalho precisa terminar em dois pontos e o corpo precisa estar endentado, ou seja, em um bloco de instruções. Por convenção, a endentação sempre é de quatro espaços. O corpo pode conter mais de uma instrução.

**Atenção:** bloco de instrução é o conjunto das instrução que estejam num mesmo nível de indentação, mesmo nível hierárquico.

A definição de uma função cria um objeto de função, que tem o tipo function:

```python
print(print_refrao)
```

A sintaxe para chamar a nova função é a mesma que a das funções integradas:

```python
print_refrao()
```

Uma vez que a função tenha sido definida, é possível usá-la dentro de outra função. Por exemplo, para repetir o refrao anterior, podemos escrever uma função print_hino_de_pernambuco:

```python
def print_hino_de_pernambuco():

  print('Coração do Brasil')
  print('Em teu seio corre o sangue de heróis, rubro veio')
  print('Que há de sempre o valor traduzir')
  print('És a fonte da vida e da história')
  print('Desse povo coberto de glória')
  print('O primeiro, talvez, do porvir')
  
  print_refrao()
  
  print('Esses montes, e vales, e rios')
  print('Proclamando o valor de teus brios')
  print('Reproduzem batalhas cruéis')
  print('No presente, és a guarda avançada')
  print('Sentinela indormida e sagrada')
  print('Que defende da pátria os lauréis')
  
  print_refrao()
  
  print('...')
  
  print_refrao()
```

```python
print_hino_de_pernambuco()
```

**Observação:** é preciso criar uma função antes de executá-la, ou seja, a definição de função tem que ser executada antes que ela seja chamada.

## **Fluxo de Execução**

Para garantir que uma função seja definida antes do seu primeiro uso, é preciso saber a ordem na qual as instruções serão executadas. A execução sempre inicia na primeira instrução do programa. As instruções são executadas uma após a outra, de cima para baixo.

As definições de função não alteram o fluxo de execução do programa, mas lembre-se de que as instruções dentro da função não são executadas até que a função seja chamada. Uma chamada de função é um desvio no fluxo de execução. Em vez de ir para a instrução seguinte, o fluxo salta para o corpo da função, executa as instruções lá, e então volta para continuar de onde parou.

## **Passagem por Valor**

Quando se passa uma variável a uma função, o *Python* passa a referência ao objeto ao qual a variável se refere. Assim, se o valor é imutável, a função não modifica a variável chamada. Se o valor é mutável, a função pode modificar a variável chamada. Observe:

```python
def modificar(x, y, z):
  x+=1
  y.append(42)
  z = [99] # nova referencia
  print('x',x)
  print('y',y)
  print('z',z)

a = 77   # variável imutável
b = [99] # variável mutável
c = [28]
modificar(a, b, c)

print('a', a)
print('b', b)
print('c', c)
```

## **Número de parâmetros variáveis**


É possível especifica um função que pode receber uma quantidade indefinida de argumentos. Esses parâmetros serão "empacotados" em uma tupla chamada args. No código abaixo, note que é preciso marcar a tupla *args* com um asterisco, para que *Python* saiba que deve alocar todos os parâmetros como elementos dela, na ordem que são passados.

```python
def func_args(*args):
  print(f'tipo: {type(args)} conteúdo: {args}')
  for arg in args:
    print(f'tipo: {type(arg)} conteúdo: {arg}')

func_args(1, 'A', {'valor': 10})
```

*Python* também permite que uma função seja chamada com uma quantidade indefinida de parâmetros por nome. Para isso, é preciso declarar o parâmetro *kwargs*, que irá representar um dicionário em que *Python* irá alocar todos os parâmetros passados por nome que não foram pré-definidos na chamada da função. O parâmetro *kwargs* deve ser marcado por asteriscos duplos na definição da função. Exemplo:

```python
def func_kwargs(**kwargs):
    print(f'tipo: {type(kwargs)} conteúdo: {kwargs}')
    for key, value in kwargs.items():
        print(f'atributo: {key}, valor: {value}')

func_kwargs(nome='James', sobrenome='Bond', cargo='Agente 007')
```

**Observação:** As palavras args e kwargs são usadas como convenção, ou seja, poderiam ter qualquer outro nome, por exemplo: \*tiago e **lima.

## **Funções com Resultado e Funções Nulas**

Diversas funções que já usamos, como as funções matemáticas, devolvem resultado. Outras funções, como print, executam uma ação, mas não devolvem um valor. Se você atribuir o resultado a uma variável, recebe um valor especial chamado **None**:

```python
resultado = print_2_vezes(':P')
```

```python
print(resultado)
```

O valor **None** não é o mesmo que a string 'None'. É um valor especial que tem seu próprio tipo:

```python
print(type(resultado))
```

As funções que criamos até aqui são nulas, ou seja, não devolvem um valor. Agora vejamos uma função que não possui um comando **return**:

```python
def potencia(x, y):
  return x ** y

resultado = potencia(2,3)

print(resultado)
```

Também é possível construir funções que retornam múltiplos valores em *Python*. Vejamos uma função que recebe dois inteiros e retorna o quociente e o resto da sua divisão:

```python
def div(dividendo, divisor):
  return dividendo // divisor, dividendo % divisor

resultado = div(2,3)

print(resultado)
```

Como se pode ver, os retornos da função `div` foram impressos entre parênteses, ou seja, na forma de uma tupla. Isso significa que Python emula retornos múltiplos de funções por meio do retorno de um único valor do tipo tupla. Essa abordagem permite desconstruir o retorno em múltiplas atribuições durante a chamada. Exemplo:

```python
dividendo, divisor = 2, 3
quociente, resto = div(dividendo, divisor)
print(quociente, resto)
```

## **Passando Funções como Argumentos**

Como mencionado acima, a definição de uma função cria uma variável com o mesmo nome, cujo valor é do tipo função. Portanto funções também podem ser passadas como parâmetros para outras funções. O código abaixo define uma função que recebe um valor `x` e uma função `oper`, que define que operação será realizada com o valor de `x`:

```python
def cubo(x):
    return x ** 3

def execute_opereraca(x, operacao):
    return operacao(x)

execute_opereraca(3, cubo)
```

## **O operador lambda**

Em *Python*, é possível usar o operador **lambda** para definir funções pequenas (que ocupam apenas uma linha). Isso pode ser útil para definir funções simples que não serão usadas em outros módulos ou em várias partes diferentes do código. O operador **lambda** retorna um valor do tipo função, portanto para usar a função assim definida, pode-se atribuí-la a uma variável:

```python
cubo = lambda x: x ** 3

cubo(3)
```

# **Glossário**
- Função: uma sequência nomeada de declarações que executa alguma operação. Podem receber argumentos ou produzir um resultado;
- Definição de Função: uma instrução que cria uma função nova, especificando seu nome, parâmetros e as instruções;
- Objeto da Função: um valor é criado por uma definição de função. O nome da função é uma variável que se refere a um objeto de função;
- Cabeçalho: a primeira linha de uma função;
- Corpo: a sequência de instruções dentro de uma função
- Parâmetro: um nome usado dentro de uma função para se referir ao valor passado como argumento;
- Chamada de Função: instrução que executa uma função. É composta pelo nome da função seguido de argumentos entre parênteses;
- Variável Local: variável definida dentro de uma função e por este motivo só pode ser visualizada dentro da função;
- Valor de Retorno: o resultado de uma função;
- Função com Resultado: uma função que devolve um valor;
- Função Nula: uma função que devolve None;
- None: um valor especial apresentado por funções nulas;
- Módulo: um arquivo que contém uma coleção de funções relacionadas e outras definições;
- Instrução de Importação: uma instrução que lê um módulo e cria um objeto de módulo;
- Objeto de Módulo: um valor criado por uma instrução import que oferece acesso aos valores definidos em um módulo;
- Notação de Ponto: sintaxe para chamar funções em outro módulo - o nome do módulo seguido do ponto e o nome da função;
- Fluxo de Execução: a ordem na qual as instruções são executadas.
