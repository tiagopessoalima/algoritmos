Para escrever programas úteis, quase sempre precisamos da capacidade de verificar condições e mudar o comportamento do programa de acordo com elas. Instruções condicionais nos dão esta capacidade. A forma mais simples é a instrução `if`, ilustrada abaixo:

```python
x, y = 2, 1

if x > y:
  print('x é maior que y')
```

A expressão depois do `if` é chamada de condição. Se ela for verdadeira, a instrução endentada é executada. Se não, nada acontece. Observe:

```python
if y > x:
  print('y é maior que x')
```

**Observação 1:** instruções `if` têm a mesma estrutura que definições de função: um cabeçalho seguido de um corpo endentado. 

**Observação 2:** não há limite para o número de instruções que podem aparecer no corpo, mas deve haver pelo menos uma.



## **Instrução pass**

Ocasionalmente, é útil ter um corpo sem instruções (normalmente como um espaço reservado para código que ainda não foi escrito). Neste caso, você pode fazer uso da instrução *pass*, que não faz nada.

```python
if x == y:
  pass # A FAZER: lidar com valores iguais!
```

## **Execução Alternativa**

É aquela na qual há duas possibilidades e a condição determina qual será executada. A sintaxe pode ser algo assim:

```python
if x == y:
  print('x é igual a y')
else:
  print('x é diferente de y')
```

Para este tipo de condição, o *Python* ainda oferece uma opção mais concisa:

```python
print('x é igual a y') if x == y else print('x é diferente de y')
```

Se `x` for igual a `y`, então o primeiro print será executado. Caso contrário, o segundo print será executado.

**Observação:** as alternativas são chamadas de branches, porque são ramos no fluxo da execução.

## **Condicionais Encadeadas**

Quando há mais de duas possibilidades, precisamos de mais que dois ramos. Esta é uma condicional encadeada:

```python
if x == y:
  print('x é igual a y')
elif x < y:
  print('x é menor que y')
elif x > y:
  print('x é maior que y')
```

**Observação:** `elif` é uma abreviatura de “else if”. Novamente, exatamente um ramo será executado. Não há nenhum limite para o número de instruções `elif`. Se houver uma cláusula else, ela deve estar no fim, mas não é preciso haver uma.

```python
if x == y:
  print('x é igual a y')
elif x < y:
  print('x é menor que y')
else:
  print('x é maior que y')
```

**Observação:** As condições são verificadas em ordem. Se uma delas for verdadeira, o ramo correspondente é executado e a instrução é encerrada. Mesmo se mais de uma condição for verdade, somente o primeiro ramo verdadeiro é executado.

## **Condicionais Aninhadas**

Uma condicional também pode ser aninhada dentro de outra. Poderíamos ter escrito o exemplo na seção anterior desta forma:

```python
if x == y:
  print('x é igual a y')
else:
  if x < y:
    print('x é menor que y')
  else:
    print('x é maior que y')
```

Para este tipo de condição, o *Python* ainda oferece uma opção mais concisa:

```python
print('x é igual a y') if x == y else print('x é menor que y') if x < y else print('x é maior que y')
```

**Atenção:** condicionais aninhadas são difíceis de ler rapidamente. É uma boa ideia evitá-las quando for possível. Operadores lógicos muitas vezes oferecem uma forma de simplificar instruções condicionais aninhadas. Observe o trecho de código a seguir:

```python
if 0 < x:
  if x < 10:
    print('x é um número positivo de um dígito.')
```

Podemos reescrever o código usando uma única condicional:

```python
if 0 < x and x < 10:
  print('x é um número positivo de um dígito.')
```

Para este tipo de condição, o *Python* ainda oferece uma opção mais concisa:

```python
if 0 < x < 10:
  print('x é um número positivo de um dígito.')
```

1. Faça uma função que receba três números e mostre-os em ordem crescente. 

```python
def ordem_crescente(valor1, valor2, valor3):
  
  if valor3 < valor1:
    valor1, valor3 = valor3, valor1
  if valor3 < valor2:
    valor2, valor3 = valor3, valor2
  if valor2 < valor1:
    valor1, valor2 = valor2, valor1
  
  print(valor1, valor2, valor3)

ordem_crescente(2,3,1)
ordem_crescente(3,2,1)
ordem_crescente(1,2,3)
ordem_crescente(1,3,2)
```

2. Faça uma função que receba a altura (metros) e o sexo. Então, calcule e retorne seu peso ideal, utilizando as seguintes formulas: 

- Homens: (72.7 ∗ altura) − 58
- Mulheres: (62.1 ∗ altura) − 44.7

```python
def peso_ideal(sexo, altura):
  if sexo == 'masculino':
    return 72.7 * altura - 58
  elif sexo == 'feminino':
    return 62.1 * altura - 44.7
  else:
    return -1
```

3. Leia a nota e o numero de faltas de um aluno, e crie uma função que retorne seu conceito. De acordo com a tabela abaixo, quando o aluno tem mais de 20 faltas ocorre uma redução de conceito.

NOTA | CONCEITO (ATE 20 FALTAS) | CONCEITO (MAIS DE 20 FALTAS)
--- | --- | ---
9.0 até 10.0 | A | B
7.5 até 8.9 | B | C
5.0 até 7.4 | C | D
4.0 até 4.9 | D | E
0.0 até 3.9 | E | E

```python
def conceito(nota, faltas):
  
  if nota >= 9:
    c = 'A'
  elif nota >= 7.5:
    c = 'B'
  elif nota >= 5:
    c = 'C'
  elif nota >= 4:
    c = 'D'
  else:
    c = 'E'
  
  if (faltas <= 20) or (c == 'E'):
    return c
  else:
    return chr(ord(c)+1)

conceito(4,10)
```

4. Faça uma função que calcule o IMC de um indivíduo e que retorne a sua classificação de acordo com a tabela abaixo:

IMC | Classificação
--- | ---
< 18,5 | Abaixo do Peso
18,6 - 24,9 | Saudável
25,0 - 29,9  | Peso em Excesso
30,0 - 34,9  | Obesidade Grau I
35,0 - 39,9  | Obesidade Grau II (severa)
≥ 40,0 | Obesidade Grau III (mórbida)

```python
def classificacao_imc(peso, altura):
  imc = peso/altura**2
  if imc <= 18.5:
    return 'Abaixo do peso'
  elif imc <= 24.9:
    return 'Saudável'
  elif imc <= 29.9:
    return 'Peso em excesso'
  elif imc <= 34.9:
    return 'Obesidade Grau I'
  elif imc <= 39.9:
    return 'Obesidade Grau II'
  else:
    return 'Obesidade Grau III'

classificacao_imc(96, 1.74)
```
