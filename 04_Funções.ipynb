# **Introdução**

Em programação, uma função é uma sequência nomeada de instruções que executa uma operação de computador. Ao definir uma função, especifica-se: i) o nome; e a ii) sequência de instruções. Depois disso, é possível chamá-la pelo nome quantas vezes forem necessárias.

## **Chamada de Função**

Já vimos até aqui diversos exemplos de chamadas de funções embutidas, como por exemplo:

```python
type(42)
```

O nome da função acima é *type*. A expressão entre parênteses é chamada de argumento da função. Para esta função, o resultado é o tipo do argumento. É comum dizer que uma função recebe um argumento e retorna um resultado. O resultado também é chamado de valor de retorno.

## **Funções Matemáticas**

*Python* possui diferentes módulos (ou bibliotecas) que fornecem diversas funções úteis. Um desses módulos é o *math* que fornece várias funções matemáticas prontas e que são úteis para realizar cálculos matemáticos. Para usar qualquer módulo em *Python* é necessário importá-lo antes. Por exemplo, para o módulo *math* é necessário acrescentar o seguinte comando antes de executar suas funções:

```python
import math
```

Esta instrução cria um objeto de módulo chamado *math*. Ao se exibir o objeto de módulo, são apresentadas informações sobre ele:



```python
math
```

A importação é necessária apenas uma vez para cada programa. Após o comando de importação do módulo, é possível usar as funções definidas por ele. Para acessá-las é preciso especificar o nome do módulo e o nome da função, separados por um ponto. Este formato é chamado de notação de ponto. A seguir exploraremos várias funções elementares.

### **math.sqrt(x)**

É uma função para calcular a raiz quadrada de do argumento x.

```python
resultado = math.sqrt(9)
print(resultado)
```

### **math.cos(x)**

Função para calcular o cosseno do argumento x, em radianos.

```python
resultado = math.cos(30)
print(resultado)
```

### **math.tan(x)**

Função para calcular a tangente do argumento x, em radianos.

```python
resultado = math.tan(3.1417)
print(resultado)
```

### **math.sin(x)**

Função para calcular o seno de do argumento x.

```python
resultado = math.sin(3.1417)
print(resultado)
```

# **Como Acrescentar Novas Funções?**

Até agora usamos funções que vêm do *Python*, mas também é possível acrescentar novas funções. Uma definição de função especifica o nome de uma nova função e a sequência de instruções que são executadas quando a função é chamada. Abaixo está um exemplo:

```python
def print_refrao():
  print('\nSalve, ó terra dos altos coqueiros')
  print('De belezas soberbo estendal')
  print('Nova Roma de bravos guerreiros')
  print('Pernambuco imortal, imortal!\n')
```

**def** é uma palavra-chave que indica uma definição de função. O nome da função é print_refrao. As regras para os nomes da função são as mesmas que as das variáveis: letras, números e sublinhados são legais, mas o primeiro caractere não pode ser um número.

**Atenção:** não podemos usar palavras-chave como nome de uma função e devemos evitar ter uma variável e uma função com mesmo nome.

Os parênteses vazios depois do nome indicam que esta função não usa argumentos. A primeira linha da definição de função chama-se **cabeçalho**; o resto é chamado de **corpo**. O cabeçalho precisa terminar em dois pontos e o corpo precisa estar endentado. Por convenção, a endentação sempre é de quatro espaços. O corpo pode conter mais de uma instrução.

A definição de uma função cria um objeto de função, que tem o tipo function:

```python
print(print_refrao)
```

A sintaxe para chamar a nova função é a mesma que a das funções integradas:

```python
print_refrao()
```

Uma vez que a função tenha sido definida, é possível usá-la dentro de outra função. Por exemplo, para repetir o refrao anterior, podemos escrever uma função print_hino_de_pernambuco:

```python
def print_hino_de_pernambuco():
  print('Coração do Brasil')
  print('Em teu seio corre o sangue de heróis, rubro veio')
  print('Que há de sempre o valor traduzir')
  print('És a fonte da vida e da história')
  print('Desse povo coberto de glória')
  print('O primeiro, talvez, do porvir')
  print_refrao()
  print('Esses montes, e vales, e rios')
  print('Proclamando o valor de teus brios')
  print('Reproduzem batalhas cruéis')
  print('No presente, és a guarda avançada')
  print('Sentinela indormida e sagrada')
  print('Que defende da pátria os lauréis')
  print_refrao()
  print('...')
  print_refrao()
```

```python
print_hino_de_pernambuco()
```

**Atenção:** é preciso criar uma função antes de executá-la, ou seja, a definição de função tem que ser executada antes que ela seja chamada.

## **Fluxo de Execução**

Para garantir que uma função seja definida antes do seu primeiro uso, é preciso saber a ordem na qual as instruções serão executadas. A execução sempre inicia na primeira instrução do programa. As instruções são executadas uma após a outra, de cima para baixo.

As definições de função não alteram o fluxo de execução do programa, mas lembre-se de que as instruções dentro da função não são executadas até que a função seja chamada. Uma chamada de função é um desvio no fluxo de execução. Em vez de ir para a instrução seguinte, o fluxo salta para o corpo da função, executa as instruções lá, e então volta para continuar de onde parou.

## **Parâmetros e Argumentos**

Algumas funções que vimos até aqui exigem argumentos. Por exemplo, ao chamar *math.sin*, você usa um número como argumento. Algumas funções exigem mais de um argumento: o *math.pow* exige dois, a base e o expoente. Dentro da função, os argumentos são atribuídos a variáveis chamada **parâmetros**. Abaixo é definida uma função que precisa de um argumento:

```python
def print_2_vezes(texto):
  print(texto)
  print(texto)
```

Esta função atribui o argumento a um parâmetro chamado texto. Quando a função é chamada, ela exibe o valor do parâmetro duas vezes.

```python
print_2_vezes(':-)')
```

```python
print_2_vezes(86)
```

```python
print_2_vezes(math.cos(math.pi))
```

Você também pode usar uma variável como argumento:

```python
nome = 'Tiago Pessoa Ferreira de Lima'
print_2_vezes(nome)
```

## **As variávis e os parâmetros são locais**



Quando você cria uma variável dentro de uma função, ela é **local**, ou seja, ela só existe dentro da função. Exemplo:

```python
def somar(operando1, operando2):
  resultado = operando1 + operando2
  print(resultado)
```

```python
valor1, valor2 = 2, 3
somar(valor1, valor2)
```

Quando a função somar é encerrada, a variável resultado é destruída. Se tentarmos exibí-la, recebemos uma exceção:

```python
print(resultado)
```

**Observação:** os parâmetros também são locais.

```python
print(valor1)
```

```python
print(valor2)
```

```python
print(operando1)
```

```python
print(operando2)
```

# **Por que funções?**

- Criar uma nova função dá a oportunidade de nomear um grupo de instruções, o que deixa o programa mais fácil de ler e depurar.
- Funções podem tornar um programa menor, eliminando o código repetitivo. Depois, se fizer alguma alteração, basta fazê-la num só lugar.
- Funções bem projetadas podem ser úteis para muitos programas. Ou seja, você pode reutilizá-la inclusive em outros programas.

# **Glossário**
- Função: uma sequência nomeada de declarações que executa alguma operação. Podem receber argumentos ou produzir um resultado;
- Definição de Função: uma instrução que cria uma função nova, especificando seu nome, parâmetros e as instruções;
- Objeto da Função: um valor é criado por uma definição de função. O nome da função é uma variável que se refere a um objeto de função;
- Cabeçalho: a primeira linha de uma função;
- Corpo: a sequência de instruções dentro de uma função
- Parâmetro: um nome usado dentro de uma função para se referir ao valor passado como argumento;
- Chamada de Função: instrução que executa uma função. É composta pelo nome da função seguido de argumentos entre parênteses;
- Variável Local: variável definida dentro de uma função e por este motivo só pode ser visualizada dentro da função;
- Valor de Retorno: o resultado de uma função;
- Função com Resultado: uma função que devolve um valor;
- Função Nula: uma função que devolve None;
- None: um valor especial apresentado por funções nulas;
- Módulo: um arquivo que contém uma coleção de funções relacionadas e outras definições;
- Instrução de Importação: uma instrução que lê um módulo e cria um objeto de módulo;
- Objeto de Módulo: um valor criado por uma instrução import que oferece acesso aos valores definidos em um módulo;
- Notação de Ponto: sintaxe para chamar funções em outro módulo - o nome do módulo seguido do ponto e o nome da função;
- Fluxo de Execução: a ordem na qual as instruções são executadas.
